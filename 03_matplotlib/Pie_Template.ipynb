# 円グラフの作成手順

```python
import matplotlib.pyplot as plt
import matplotlib
import japanize_matplotlib

data = [70, 15, 5, 10]
labels = ["株式", "債券", "現金", "不動産"]

plt.pie(data,labels=labels,counterclock=False, startangle=90,wedgeprops={'linewidth': 3, 'edgecolor':"white"},autopct="%1.0f%%")

plt.axis("equal")
#plt.savefig('image.jpg',figsize=(150, 150), dpi=300)
```
